```python
def fileOperation(file):
    upper_count=0
    file_string=''
    outfile= open("result.txt","w")#output file
    for filename in filenames:#reading eachfile
        file = open(filename, "r")
        for line in file:
            file_string+=line #concatinating each line to string
        for char in file_string:
            if char.isupper():
                upper_count+=1 #counting upper case character
        wordlist=file_string.split() #split the words
        wordfreq =[]
        wordlistUniq=[]
        for word in wordlist:
            if word not in wordlistUniq:
                wordlistUniq.append(word) #storing unique word
        for word in wordlistUniq:
            wordfreq.append(wordlist.count(word)) #counting frequencies
        file.close()#closing current file
        outfile.write('Number of upper case character in '+filename+' : ' +str(upper_count)+"\n") #wrinting upper count to file
        outfile.write('Word frequencies\n')
        for word,count in zip(wordlistUniq,wordfreq):
            outfile.write(word+" : " +str(count)+'\n') #writing word frequencies
        outfile.write('\n')
        file_string=''    
    outfile.close() #closing output file

filenames=['untitled.txt'] #for testing
fileOperation(filenames)

```
